# Area and Population Analysis

Here we did simple EDA and plot the data in an understandable way. Here is the 
<a href=https://github.com/amankharwal/Californai-Cities/blob/master/california_cities.csv>dataset source</a>

```python
import pandas as pd
```

```python
# EDA - exploring the data
```

```python
df = pd.read_csv('california_cities.csv')
```

```python
df['area_land_km2']=df['area_land_km2'].fillna(0)
```

```python
df['area_land_km2'].isnull().sum()
```

```python
# extracting useful data
```

```python
latitude, longitude = df['latd'], df['longd']
```

```python
population, area = df['population_total'], df['area_total_km2']
```

```python
# Plotting the scatter plot
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
plt.scatter(longitude, latitude, label=None, c=np.log10(population),
            cmap='viridis', s=area, linewidth=0, alpha=0.5)
plt.axis('equal')
plt.xlabel('Longitude')
plt.ylabel('Longitude')
plt.colorbar(label='log$_{10}$(population)')
plt.clim(3, 7)
#Legend created. here we created a null scatter plot. it is just to show the area legend seperatly.
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')
    plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.title("Area and Population of California Cities")
plt.show()
```
